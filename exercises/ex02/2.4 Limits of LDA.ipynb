```python
def estimate_covariance(X):
    '''estimate the covariance matrix
    Input: data matrix X (size: NxD) 
    Output: covariance matrix cov (size: ?x?)'''
    
    return 1/(X.shape[0] - 1) * np.dot((X - np.mean(X, 0)).T, (X - np.mean(X, 0)))

    

def train_lda(X, y):
    ''' Train an LDA
    Input: X data matrix with shape NxD
           y label vector with shape Nx1
    Output: weight vector with shape Nx1 
            bias term - real-valued '''
    
    # calc class means
    m1 = 1/len(np.where(y == 0)[0]) * (np.sum(X[np.where(y == 1)[0]], axis=0).T)
    m2 = 1/len(np.where(y == 1)[0]) * (np.sum(X[np.where(y == 2)[0]], axis=0).T)

    # calc S_w_inverted
    S_1 = estimate_covariance(X[np.where(y == 1)[0]])
    S_2 = estimate_covariance(X[np.where(y == 2)[0]])
    
    S_w = 1/2 * np.add(S_1, S_2)
    S_w_inverted = np.linalg.pinv(S_w)
    
    weights = np.dot(S_w_inverted, np.subtract(m1, m2))
    bias = -1/2 * np.dot(weights, np.add(m1, m2)) 
    
    return weights, bias

def apply_lda(X_test, weights, bias):
    '''Predict the class label per sample 
    Input: X_test - data matrix with shape NxD
           weight vector and bias term from train_LDA
    Output: vector with entries 1 or 2 depending on the class''' 
    y_test = np.array([])
    for x in X_test:
        
        val = np.dot(weights, x) + bias

        y = 1
        if val < 0:
            y = 2
        y_test = np.append(y_test, y)

    return y_test
```

```python

ds_data = 'data/unbalanced_train.data'
ds_labels ='data/unbalanced_train.labels'
X = np.loadtxt(ds_data)
y = np.loadtxt(ds_labels)

# 1-> Scatter-plot your 2D-data.
c1 = X[np.where(y == 0)[0]]
c2 = X[np.where(y == 1)[0]]


plt.scatter(c1[:, 0], c1[:, 1], c="b")
plt.scatter(c2[:, 0], c2[:, 1], c="g")

plt.grid()
plt.show()

# 2-> What is the optimal decision boundary?

w = np.dot(np.dot(np.linalg.pinv(np.dot(X.T, X)), X.T), y)

def h(x):
    return 1/(1 + math.e ** (- np.dot(w.T, x)))

y_prediction = np.array([])
for x in X:
    l = 0
    if h(x) >= 0.5:
        l = 1
    y_prediction = np.append(y_prediction, l)

print(sum(y == y_prediction)/len(y_prediction))
```
